    So we use Twint package that can help us scrape Twitter data without Twitter Developer Account.


```python
import twint
import nest_asyncio
nest_asyncio.apply()
```

# Get CDC tweets containing COVID

```python
# Configure
c = twint.Config()

c.Username = "CDCgov"
c.Search = "COVID"
c.Store_json = True
c.Output = "data\cdc_twitter_covid.json"

# Run
twint.run.Search(c)
```

# Get All CDC Tweets

```python
# Configure
c = twint.Config()

c.Username = "CDCgov"
# c.Search = "COVID"
c.Since = '2020-01-01'
c.Store_json = True
c.Output = "data\cdc_twitter_since_2020.json"

# Run
twint.run.Search(c)
```
