```python
import pandas as pd

#Import cleaned data
tweets_df = pd.read_csv("tweet.csv")
covid_us_df = pd.read_csv("covid-data-US.csv")
covid_state_df = pd.read_csv("covid-data-by-state.csv")
covid_state_df = pd.read_csv("covid-data-by-state.csv")
```

```python
#Minor preprocessing
print(tweets_df.isna()
        .sum())
tweets_df.fillna(0, inplace = True)
tweets_df.head()
```

```python
tweets_df = tweets_df.drop(columns=tweets_df.columns[0])
tweets_df.head()
```

```python
print(covid_us_df.isna()
        .sum())
covid_us_df.fillna(0, inplace = True)
covid_us_df.head()
```

```python
print(covid_state_df.isna()
        .sum())
covid_state_df.fillna(0, inplace = True)
covid_state_df.head()
```

```python
#Linkage
merged_df = pd.merge(left = tweets_df, right = covid_us_df, on = 'date')
merged_df = pd.merge(left = merged_df, right = covid_state_df, on = 'date')
```

```python
print(covid_state_df.isna()
        .sum())
merged_df.head()
```

```python
#Output merged data to CSV
merged_df.to_csv("merged_us_state_tweet.csv", sep='\t', encoding='utf-8')
```
